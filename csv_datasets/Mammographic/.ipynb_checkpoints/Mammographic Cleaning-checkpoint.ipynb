```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mammographic_masses_original.csv', na_values=['#NAME ?'])
data.head()
```

```python
data.info()
```

```python
data = data[data != '?']
data.info()
```

```python
data = data.dropna()
data.info()
```

```python
data = data.astype('int64')
```

```python
corrmat = data.corr()
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, vmax=1, square=True);
```

```python
k = 6
cols = corrmat.nlargest(k, 'Severity')['Severity'].index
cm = np.corrcoef(data[cols].values.T)
f, ax = plt.subplots(figsize=(10, 10))
sns.set(font_scale=2)
sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 20}, yticklabels=cols.values, xticklabels=cols.values)
```

```python
#remove correlation coef below 0.2
data = data.drop(columns = 'Density')
data.head()
```
